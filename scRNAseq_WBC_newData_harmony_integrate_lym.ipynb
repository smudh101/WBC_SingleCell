```r
str(total.lym.integrated)
```

```r
exp.M=as.matrix(total.lym.integrated@assays$RNA@data)
```

```r
lym <- CreateSeuratObject(exp.M)
```

```r
lym <- SetAssayData(lym, "data", exp.M)
```

```r
lym <- AddMetaData(lym, total.lym.integrated@meta.data)
```

```r
table(lym$new.batch)
```

```r
table(lym$batch)
```

```r
lym <- FindVariableFeatures(lym, selection.method = "vst", nfeatures = 2000)
lym <- ScaleData(lym, verbose = FALSE)
lym <- RunPCA(lym, npcs = 30, verbose = FALSE)
```

```r
options(repr.plot.height = 5, repr.plot.width = 18)
p1 <- DimPlot(object = lym, reduction = "pca", pt.size = .1, group.by = "batch")
p2 <- VlnPlot(object = lym, features = "PC_1", group.by = "batch", pt.size = .1)
plot_grid(p1,p2)
```

```r
lym=RunHarmony(lym,"batch", plot_convergence = TRUE)
```

```r
harmony_embeddings <- Embeddings(lym, 'harmony')
harmony_embeddings[1:5, 1:5]
```

```r
options(repr.plot.height = 5, repr.plot.width = 18)
p1 <- DimPlot(object = lym, reduction = "harmony", pt.size = .1, group.by = "batch")
p2 <- VlnPlot(object = lym, features = "harmony_1", group.by = "batch", pt.size = .1)
plot_grid(p1,p2)
```

```r
lym <- RunUMAP(lym, reduction = "harmony", dims = 1:30)
lym <- FindNeighbors(lym, dims = 1:30,reduction = "harmony")
lym <- FindClusters(lym, resolution = 0.5)
```

```r
options(repr.plot.width=24,repr.plot.height=4)
DimPlot(lym,reduction = "umap",label = T,label.size = 6,split.by = "batch",group.by = "cellAnno",cols = usedCols)
```
