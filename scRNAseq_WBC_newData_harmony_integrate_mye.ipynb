```r
str(total.mye.integrated)
```

```r
exp.M=as.matrix(total.mye.integrated@assays$RNA@data)
```

```r
mye <- CreateSeuratObject(exp.M)
```

```r
mye <- SetAssayData(mye, "data", exp.M)
```

```r
mye <- AddMetaData(mye, total.mye.integrated@meta.data)
```

```r
table(mye$new.batch)
```

```r
table(mye$batch)
```

```r
mye <- FindVariableFeatures(mye, selection.method = "vst", nfeatures = 2000)
mye <- ScaleData(mye, verbose = FALSE)
mye <- RunPCA(mye, npcs = 30, verbose = FALSE)
```

```r
options(repr.plot.height = 5, repr.plot.width = 18)
p1 <- DimPlot(object = mye, reduction = "pca", pt.size = .1, group.by = "batch")
p2 <- VlnPlot(object = mye, features = "PC_1", group.by = "batch", pt.size = .1)
plot_grid(p1,p2)
```

```r
mye=RunHarmony(mye,"batch", plot_convergence = TRUE)
```

```r
harmony_embeddings <- Embeddings(mye, 'harmony')
harmony_embeddings[1:5, 1:5]
```

```r
options(repr.plot.height = 5, repr.plot.width = 18)
p1 <- DimPlot(object = mye, reduction = "harmony", pt.size = .1, group.by = "batch")
p2 <- VlnPlot(object = mye, features = "harmony_1", group.by = "batch", pt.size = .1)
plot_grid(p1,p2)
```

```r
mye <- RunUMAP(mye, reduction = "harmony", dims = 1:30)
mye <- FindNeighbors(mye, dims = 1:30,reduction = "harmony")
mye <- FindClusters(mye, resolution = 0.5)
```

```r
options(repr.plot.width=24,repr.plot.height=4)
DimPlot(mye,reduction = "umap",label = T,label.size = 6,split.by = "batch",group.by = "cellAnno")
```

```r
options(repr.plot.width=7,repr.plot.height=6)
DimPlot(mye,reduction = "umap",label = T,label.size = 6,group.by = "cellAnno",cols = usedCol)
```
